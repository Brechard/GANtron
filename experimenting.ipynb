Width of the guided attention

```python
from loss_function import attention_mask
import matplotlib.pyplot as plt

figure_path = 'reports/figures/'
```

```python
mask = attention_mask(n_chars=40, n_frames=240, width=0.5).cpu().numpy()
plt.imshow(mask, aspect='auto', origin='lower', interpolation='none')
plt.title("Thing attention guidance")
plt.savefig(figure_path + 'thing_guidance.png')
plt.show()
```

```python
mask = attention_mask(n_chars=40, n_frames=240, width=2).cpu().numpy()
plt.imshow(mask, aspect='auto', origin='lower', interpolation='none')
plt.title("Wide attention guidance")
plt.savefig(figure_path + 'wide_guidance.png')
plt.show()
```
